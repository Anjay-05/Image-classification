```python
#importing libraries
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import requests
```

```python
#Connecting to web driver
driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")

#Getting the web page
url="https://www.amazon.in"
driver.get(url)
# Products to be search
product=['Sarees','Mens Trousers','Mens Jeans']

for item in product:
    #Entering the data to be search
    search=driver.find_element_by_id("twotabsearchtextbox")
    search.clear()
    search.send_keys(item)

    #Clicking search button
    searchbutton=driver.find_element_by_id("nav-search-submit-button")
    searchbutton.click()
    
    

    img_urls = []
    img_data = []

    for k in range(4):
        Page=[]
        page=driver.find_elements_by_xpath('//span[@class="s-pagination-strip"]/a')#Getting the next page url
        for i in page:
            Page=(i.get_attribute('href'))
        if len(img_urls)<=200:
            images=driver.find_elements_by_xpath('//div[@class="a-section aok-relative s-image-tall-aspect"]/img')
            for i in images:
                source= i.get_attribute('src')
                if source is not None:
                    if(source[0:4] == 'http'):
                        img_urls.append(source)

        print(Page)
        time.sleep(5)
        driver.get(Page)


    for j in range(len(img_urls)):
        if j >= 200:
            break
        print("Downloading {0} of {1} images" .format(j, 200))
        response= requests.get(img_urls[j])
        file = open(r"C:\Users\user\OneDrive\Pictures\images\{}".format(item)+str(j)+".jpg", "wb")
        file.write(response.content)
        
```
